# Nivel de estudios de los usuarios.

```python
personas = personas.rename(index=str,columns={"nombre": "estudios"})
```

```python
personas.groupby('estudios').count().sort_values(by='idpostulante').loc[:,'idpostulante'].plot.bar()
```

# Estudios de los usuarios segun su genero.

```python
personas.groupby(['estudios','sexo']).count().loc[:,['idpostulante']].plot.bar()
```

```python

aux = personas
aux['FEM'] = personas['sexo'] == 'FEM'
aux['MASC'] = personas['sexo'] == 'MASC'
aux.loc[personas['sexo'] == 'FEM','MASC'] = np.NaN
aux.loc[personas['sexo'] == 'MASC','FEM'] = np.NaN
plt = aux.groupby('estudios').count().sort_values('idpostulante')[-20:].loc[:,['MASC','FEM']].plot.bar(figsize=(15,12))
#plt.set_title('Cantidad de avisos por areas separado en Full-time y Part-time', size=20)
#plt.set_xlabel('Area de trabajo',size=20)
#plt.set_ylabel('Cantidad de avisos',size=20)
plt.tick_params(axis='x', labelrotation=70, labelsize=15)
plt.tick_params(axis='y', labelsize=15)
```

# Estado de los estudios de los usuarios.

```python
personas.groupby(['estudios','estado']).count().loc[:,['idpostulante']].plot(kind='bar')
```

```python
personas.groupby(['estudios','estado']).count().loc[:,['idpostulante']]
```

```python
personas.groupby(['estudios','estado']).count().loc[:,['idpostulante']].sort_values(by='idpostulante').plot.bar()
```

# Viendo mejor los datos anteriores...
### Como la gran mayoria de los usuarios tiene terminado el secundario, podria ser que este no sea un dato relevante para clasificar personas. Por lo tanto los sacamos momentaneamente y volvemos a graficar el estado de los estudios.

```python
personas[(personas.estudios!="Secundario")&(personas.estado!='Graduado')].groupby(['estudios','estado']).count().loc[:,['idpostulante']].sort_values(by='idpostulante').plot.bar()
```
